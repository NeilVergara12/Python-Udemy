MAP FUNCTION

```python
def square(num):
    return num**2
```

```python
my_nums = [1,2,3,4,5]
```

```python
map(square,my_nums)
```

```python
list(map(square,my_nums))
```

```python
for item in map(square,my_nums):
    print(item)
```

```python
list(map(square,my_nums))
```

```python
def splicer(mystring):
    if len(mystring)%2 == 0:
        return 'EVEN'
    else:
        return mystring[0]
```

```python
names = ['Andy','Eve','Sally']
```

```python
list(map(splicer,names))
```

FILTER FUNCTION

```python
def check_even(num):
    return num % 2 == 0 
```

```python
mynums = [1,2,3,4,5,6]
```

```python
list(filter(check_even,mynums))
```

```python
for n in filter(check_even,mynums):
    print(n)
```

LAMBDA ESPRESSION

```python
def square(num):
    result = num**2
    return result
```

```python
square(3)
```

```python
def square(num):
    return num**2
```

```python
square(3)
```

```python
square = lambda num: num**2
```

```python
square(5)
```

```python
list(map(lambda num:num**2,mynums))
```

```python
list(filter(lambda num:num % 2 == 0,mynums))
```
