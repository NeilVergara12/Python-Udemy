LEGB

L-local
E-Enclosing function locals
G-Global(Module)
B-Built-in(Python)

```python
name = 'THIS IS A GLOBAL STRING'

def greet():
    
    name = 'Neil'
    
    def hello():
        print('Hello '+name)
    
    hello()

greet()
```

```python
greet()
```

```python
name = 'THIS IS A GLOBAL STRING'

def greet():
    
    # name = 'Neil'
    
    def hello():
        print('Hello '+name)
    
    hello()

greet()
```

```python
name = 'THIS IS A GLOBAL STRING'

def greet():
    # ENCLOSING
    name = 'Neil'
    
    def hello():
        # LOCAL
        name = 'IM A LOCAL'
        print('Hello '+name)
    
    hello()

greet()
```

```python
x = 50

def func(x):
    print(f'X is {x}')

    # LOCAL REASSIGNMENT
    x = 200
    print(f'I JUST LOCALLY CHANGE X TO {x}')
```

```python
func(x)
```

```python
x = 50

def func():
    global x
    print(f'X is {x}')

    # LOCAL REASSIGNMENT ON GLOBAL VARIABLE
    x = 'NEW VALUE'
    print(f'I JUST CHANGE GLOBAL X TO {x}')
```

```python
print(x)
```

```python
func()
```

```python
print(x)
```
