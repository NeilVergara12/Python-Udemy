```python
def func_one(n):
    return[str(num) for num in range(n)]
```

```python
func_one(10)
```

```python
def func_two(n):
    return list(map(str,range(n)))
```

```python
func_two(10)
```

```python
import time
```

```python
#CURRENT TIME BEFORE
start_time = time.time()
#RUN CODE
result = func_one(1000000)
#CURRENT TIME AFTER RUNNING CODE
end_time = time.time()
#ELAPSED TIME
elapsed_time = end_time - start_time

print(elapsed_time)
```

```python
#CURRENT TIME BEFORE
start_time = time.time()
#RUN CODE
result = func_two(1000000)
#CURRENT TIME AFTER RUNNING CODE
end_time = time.time()
#ELAPSED TIME
elapsed_time = end_time - start_time

print(elapsed_time)
```
