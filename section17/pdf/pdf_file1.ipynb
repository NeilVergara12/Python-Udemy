```python
f = open('Working_Business_Proposal.pdf','rb')

pdf_text = []

pdf_reader = pypdf.PdfReader(f)

for num in range(pdf_reader.get_num_pages()):

    page = pdf_reader.get_page(num)
    pdf_text.append(page.extract_text())
```
